```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Loading the data.
data = pd.read_csv('breast-cancer-wisconsin-data/data.csv')
```

```python
# Checking out how the data looks.
data.head(5)
```

```python
#Getting some global data of the data set.
data.info()
```

```python
#Identifing those features which are null.
data.isnull()
```

```python
#We can actually check the null features better using a heatmap plot from seaborn library.
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis');
```

```python
#Dropping the feature Unnamed: 32 since it's the only null feature.
data.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
#Plotting the heatmap again. Just in case.
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis');
```

```python
#Setting a specific style for the graphs.
sns.set_style('whitegrid')
```

```python
#Plotting the amount of benign and malign examples we have in our data set.
sns.countplot(x='diagnosis',data=data,palette='RdBu_r');
```

```python
#Checking the distribution of the feature radius mean. It looks like a combination of two Gaussians.
#It makes sense the largest one has a lower mean (more benign cases) and the shortest one has a higher mean (less maling cases).
sns.distplot(data['radius_mean']);
```

```python
#With this plotting we can also see the standar desviation.
sns.boxplot(x='diagnosis',y='radius_mean',data=data);
```

```python
#Checking the distribution of some features and their correlation for benign cases (orange) and malign cases (blue).
sns.pairplot(data=data, hue='diagnosis',
             x_vars=['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean'],
             y_vars=['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean'])
```

```python
#Since we want numerical data, we change the feature diagnosis values M and B for 0 and 1.
data.loc[data['diagnosis']=='M','diagnosis'] = 0
data.loc[data['diagnosis']=='B','diagnosis'] = 1
```

```python
#Checking out the results.
data.tail(5)
```

```python
#Dropping the feature id. It's not actually a feature.
data.drop('id',axis=1,inplace=True)
```
